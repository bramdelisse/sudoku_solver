```python
s = [
    [5,3,0,0,7,0,0,0,0], #R1
    [6,0,0,1,9,5,0,0,0], #R2
    [0,9,8,0,0,0,0,6,0], #R3
    [8,0,0,0,6,0,0,0,3], #R4
    [4,0,0,8,0,3,0,0,1], #R5
    [7,0,0,0,2,0,0,0,6], #R6
    [0,6,0,0,0,0,2,8,0], #R7
    [0,0,0,4,1,9,0,0,5], #R8
    [0,0,0,0,8,0,0,7,9]  #R9
]
```

```python
# print sudoku game
def print_sudoku(s=s):
    hor_line = '-'*25
    print(hor_line)
    for r in range(9):
        print('|', end=' ')
        for c in range(9):
            if s[r][c] == 0:
                print('.', end=' ')
            else:
                print(s[r][c], end=' ')
            if c%3 == 2:
                print('|', end=' ')
        print()
        if r%3 == 2:
            print(hor_line)

print_sudoku(s)
```

```python
# create function to test a cell for a number

def find_square(x):
    count = 0
    while x >= 0:
        x -= 3
        count += 1
    return [3*(count-1), 3*(count-1)+1, 3*(count-1)+2]

def test_number(guess, r_i, c_i, s=s):
    # test if cell is empty
    if s[r_i][c_i] != 0:
        print(f"🛑 number is already known: {s[r_i][c_i]}")
        return False

    # test row
    for c in s[r_i]:
        if c == guess:
            print("❌ number is present in row")
            return False

    # test column
    c_to_check = [row[c_i] for row in s]
    if guess in c_to_check:
        print("❌ number is present in column")
        return False
    else:
        pass

    # test square
    r_sq = find_square(r_i)
    c_sq = find_square(c_i)

    for r in r_sq:
        for c in c_sq:
            if s[r][c] == guess:
                print("❌ number is present in square")
                return False
            
    print("Number is possible!")
    return True
```

```python
# create function to reveal possibilities for cell

def reveal_cell(r_i, c_i, s=s):
    poss = set(range(1,10))
    
    # test row
    poss -= set(s[r_i])

    # test column
    c_to_reveal = [row[c_i] for row in s]
    poss -= set(c_to_reveal)

    # test square
    r_sq = find_square(r_i)
    c_sq = find_square(c_i)
    for r in r_sq:
        for c in c_sq:
            poss.discard(s[r][c])

    return poss
```

```python
# Iteratively solve sudoku

def verify_complete(s):
    for r in range(9):
        for c in range(9):
            if s[r][c] == 0:
                return False
    return True

def solve_sudoku(s):
    turn = 0
    while verify_complete(s) == False:
        print()
        print("="*7, f"|TURN: {turn}|", 7*"=")
        print_sudoku(s)
        # Reveal known cells
        for r in range(9):
            for c in range(9):
                if s[r][c] == 0 and len(reveal_cell(r,c,s)) == 0:
                    raise Exception(f"No number is possible in cell {(r,c)}: {s[r][c]}, mistakes have been made")
                if s[r][c] == 0 and len(reveal_cell(r,c,s)) == 1:
                    s[r][c] = reveal_cell(r,c,s).pop()
                    print(f"Row: {r}, Column: {c} -> {s[r][c]}")
        turn += 1
    
    print()
    print(f"🎉 Sudoku completed in {turn} turns")
    print_sudoku(s)
```

```python
solve_sudoku(s)
```
